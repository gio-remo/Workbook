```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
terna = yf.download("TRN.MI", start="2015-01-01", end="2023-12-31")
```

```python
terna
```

```python
plt.plot(terna.Close, color="green")
plt.title("Terna Spa")
plt.ylabel("Price €")
plt.grid()
```

```python
enel = yf.download("ENEL.MI", "2015-01-01", "2023-12-31")
```

```python
plt.plot(enel.Close, color="red", linewidth="0.8")
plt.title("Enel Spa")
plt.ylabel("Price €")
plt.grid(axis = "y")
plt.yticks(np.arange(3,9.5,0.5))
```

```python
plt.plot(terna.Close, label="Terna Spa")
plt.plot(enel.Close, label="Enel Spa")
plt.legend(loc="upper left")
plt.show()
```
